Ejercicio 7

Para determinar la temperatura crítica de la transición de fase de manera más precisa, se estudiará la separación
cuadrática media de un par de átomos, $⟨(Δr_{ij}(t))^2⟩ = ⟨(r_{i}(t) − r_{j}(t))^2⟩$ como función de la temperatura. Para
ello se calentará el sistema lentamente, es decir, se rescalarán las velocidades de las partículas por un factor de
1.1 en los tiempos t = 500, 1000, 2000, .... La temperatura crítica estará alrededor de aquel valor para el que
los átomos experimenten un alejamiento considerable y abrupto de sus posiciones.

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargo los datos 
r_cuadradito_medio_particulas = np.loadtxt("Separacion_cuadratica_media_entre_particulas.txt")
Velocidades = np.loadtxt("Velocidades_atomos_ej_7.txt")

# Paso temporal y la longitud de la caja
h = 0.002
L = 12.0


# Realizo los promedios en los intervalos de tiempo correspondientes a los reescalados de las velocidades, es decir, a t = 500, 1000, 2000 y 3000 segundos
r_cuadrado_0 = np.mean(r_cuadradito_medio_particulas[:250000])
r_cuadrado_500 = np.mean(r_cuadradito_medio_particulas[250000:500000])
r_cuadrado_1000 = np.mean(r_cuadradito_medio_particulas[500000:1000000])
r_cuadrado_2000 = np.mean(r_cuadradito_medio_particulas[1000000:1500000])
r_cuadrado_3000 = np.mean(r_cuadradito_medio_particulas[1500000:1750000])

# Almaceno en un array
r_cuadrado_medio = np.array([r_cuadrado_0, r_cuadrado_500, r_cuadrado_1000, r_cuadrado_2000, r_cuadrado_3000])

# Extraigo las velocidades 
v_x = Velocidades[0:, 0::2]
v_y = Velocidades[0:, 1::2]

# Temperatura en cada instante de tiempo 
T = 0.5 * np.mean(v_x**2 + v_y**2, axis = 1)

# Hago la media sobre los intervalos de tiempo correspondientes a los reescalados de las velocidades, es decir, a t = 500, 1000, 2000 y 3000 segundos
T_0 = np.mean(T[:250000])
T_500 = np.mean(T[250000:500000])
T_1000 = np.mean(T[500000:1000000])
T_2000 = np.mean(T[1000000:1500000])
T_3000 = np.mean(T[1500000:1750000])

# Almaceno en un array
T_medio = np.array([T_0, T_500, T_1000, T_2000, T_3000])

print ("Temperaturas medias: ", T_medio)
print ("Separaciones cuadráticas medias: ", r_cuadrado_medio)

# Grafico la evolución de r_cuadrado media en función de la temperatura 
fig, ax = plt.subplots(figsize=(5, 5))

ax.plot(T_medio, r_cuadrado_medio, marker = "o", color = 'red')
ax.set_xlabel("T", fontsize=18)
ax.set_ylabel(r"$\langle(r_i(t) - r_j(t))^2 \rangle$", fontsize=18)
ax.tick_params(axis="both", labelsize=18)

fig.savefig("ej_7.png", dpi = 150, bbox_inches="tight")
plt.show()
```
